# Check Point Source Masks

```python
testfiles = random.sample([kk for kk in os.listdir('/srv/two/zhutchen/g3rassimages_mosaics_broad/') if 'Cnt' in kk],k=5)
print(testfiles)
```

```python
mosapath = '/srv/two/zhutchen/g3rassimages_mosaics_broad/'
maskpath = '/srv/two/zhutchen/g3rassimages_masked_broad/'
fig,axs=plt.subplots(ncols=2,nrows=len(testfiles),figsize=(5,12))
for ii,ff in enumerate(testfiles):
    mosaic = fits.open(mosapath+ff)[0].data
    masked = fits.open(maskpath+ff)[0].data
    axs[ii,0].imshow(mosaic,vmin=0,vmax=5*np.mean(mosaic))
    axs[ii,1].imshow(masked,vmin=0,vmax=4*np.mean(mosaic))
axs[0,0].set_title("Original Mosaic")
axs[0,1].set_title("Masked Image")
plt.tight_layout()
#plt.savefig("../figures/masking_example.jpg",dpi=300)
plt.show()
```

```python
%matplotlib inline
testfiles = ['RASS-Cnt_Broad_grp13.0.fits']
mosapath = '/srv/two/zhutchen/g3rassimages_mosaics_broad/'
maskpath = '/srv/two/zhutchen/g3rassimages_masked_broad/'
scalpath = '/srv/two/zhutchen/g3rassimages_scaled_broad/'
fig,axs=plt.subplots(ncols=3,nrows=len(testfiles),figsize=doublecolsize)
for ii,ff in enumerate(testfiles):
    mosaic = fits.open(mosapath+ff)[0].data
    masked = fits.open(maskpath+ff)[0].data
    scaled = fits.open(scalpath+ff)[0].data
    axs[0].imshow(mosaic,vmin=0,vmax=5*np.mean(mosaic))
    axs[1].imshow(masked,vmin=0,vmax=4*np.mean(mosaic))
    axs[2].imshow(scaled,vmin=0,vmax=4*np.mean(mosaic))
#axs[0].set_title("Original Mosaic")
#axs[1].set_title("Masked Image")
#axs[2].set_title("Scaled & Cropped")
axs[0].text(30,50,'(a) Custom Mosaic',backgroundcolor='white',fontsize=8)
axs[1].text(30,50,'(b) Masked Image',backgroundcolor='white',fontsize=8)
axs[2].text(30*(220/512),50*(220/512),'(c) Scaled & Cropped',backgroundcolor='white',fontsize=8)
for ii in range(len(axs)):
    axs[ii].set_xlabel('pixels')
    if ii!=2:
        axs[ii].set_xticks(np.arange(0,600,100))
        axs[ii].set_yticks(np.arange(0,600,100))
axs[0].set_ylabel("pixels")
axs[2].set_xticks(np.arange(0,250,50))
axs[2].set_yticks(np.arange(0,250,50))
plt.tight_layout()
#plt.savefig("../figures/coma_masking_example.jpg",dpi=300)
plt.show()
```

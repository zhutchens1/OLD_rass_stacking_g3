```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import os
import random
```

# Check Point Source Masks

```python
testfiles = random.sample([kk for kk in os.listdir('/srv/two/zhutchen/g3rassimages_mosaics_broad/') if 'Cnt' in kk],k=5)
print(testfiles)
```

```python
mosapath = '/srv/two/zhutchen/g3rassimages_mosaics_broad/'
maskpath = '/srv/two/zhutchen/g3rassimages_masked_broad/'
fig,axs=plt.subplots(ncols=2,nrows=len(testfiles),figsize=(5,12))
for ii,ff in enumerate(testfiles):
    mosaic = fits.open(mosapath+ff)[0].data
    masked = fits.open(maskpath+ff)[0].data
    axs[ii,0].imshow(mosaic,vmin=0,vmax=5*np.mean(mosaic))
    axs[ii,1].imshow(masked,vmin=0,vmax=4*np.mean(mosaic))
axs[0,0].set_title("Original Mosaic")
axs[0,1].set_title("Masked Image")
plt.tight_layout()
#plt.savefig("../figures/masking_example.jpg",dpi=300)
plt.show()
```
